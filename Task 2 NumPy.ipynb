#### 24. Есть генератор, сделать с его помощью массив

```python
def countdown(num):
     while num > 0:
         yield num
         num -= 1

val = countdown(5)
a = []

for i in range(5):
  a.append(next(val))

a = np.array(a)
a
```

#### 32. Из двух массивов сделать матрицу Коши C (Cij = 1/(xi - yj))

```python
a = np.array([1, 3, 2, 1])
b = np.array([1, 1, 4, 8, 8])

t_a = np.array([i for i in a] * b.shape[0]).reshape(a.shape[0], b.shape[0])
t_b = np.array([i for i in b] * a.shape[0]).reshape(b.shape[0], a.shape[0]).T

C = 1 / ((t_a + t_b) + 1e-10)
C
```
